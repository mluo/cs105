```python
%matplotlib inline

import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import operator
import re

pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 30)

# set some nicer defaults for matplotlib
from matplotlib import rcParams

#these colors come from colorbrewer2.org. Each is an RGB triplet
dark2_colors = [(0.23137254902, 0.34901960784, 0.59607843137),
                (0.10588235294117647, 0.6196078431372549, 0.4666666666666667),
                (0.8509803921568627, 0.37254901960784315, 0.00784313725490196),
                (0.4588235294117647, 0.4392156862745098, 0.7019607843137254),
                (0.9058823529411765, 0.1607843137254902, 0.5411764705882353),
                (0.4, 0.6509803921568628, 0.11764705882352941),
                (0.9019607843137255, 0.6705882352941176, 0.00784313725490196),
                (0.6509803921568628, 0.4627450980392157, 0.11372549019607843),
                (0.4, 0.4, 0.4)]

rcParams['figure.figsize'] = (20, 16)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.grid'] = False
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 25
rcParams['patch.edgecolor'] = 'none'


def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecessary plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

```python
def rgb_friend(r, g, b):
    """ Output rgb color based on triplet values """
    return (float(r) / 255., float(g) / 255, float(b) / 255)

dark = rgb_friend(59, 89, 152)
medium = rgb_friend(95, 120, 171)
light = rgb_friend(138, 156, 194)
filler = rgb_friend(235, 239, 247)

# used to create inverse graphs for presentation
dark_inverse = rgb_friend(196, 166, 103)
medium_inverse = rgb_friend(160, 135, 84)
light_inverse = rgb_friend(117, 99, 61)
filler_inverse = rgb_friend(20, 16, 8)
```

```python
# import data

stranger_data = pd.read_csv('stranger_data.csv').set_index('Student Name')
stranger_subset = (stranger_data[stranger_data['Permission'] == 1]).drop(['Permission'], 1)
stranger_data = stranger_data.drop(['Permission'], 1)

mutual_friends_data = pd.read_csv('mutual_friends_data.csv').set_index('Student Name')
mutual_friends_data = (mutual_friends_data[mutual_friends_data['Permission'] == 1]).drop(['Permission'], 1)

friends_data = pd.read_csv('friends_data.csv').set_index('Student Name')
friends_data = (friends_data[friends_data['Permission'] == 1]).drop(['Permission'], 1)
```

```python
"""
Function
--------
plot_data_percents

Plots percentage of individuals willing to share each type of information

Parameters
----------
data : DataFrame
    Data to use
title : string
    The title information for the plot
"""
def plot_data_percents(data, title, filename=None):
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    
    data_counts = {name:len(data[data[name] != 'Unlisted']) for name in data.columns}
    sorted_data_counts = sorted(data_counts.iteritems(), key=operator.itemgetter(1))

    ax.bar(np.arange(len(sorted_data_counts)),
           [float(item[1]) / len(data.index) * 100 for item in sorted_data_counts],
           align='center', color=dark, edgecolor='white')

    ax.set_xticklabels([item[0] for item in sorted_data_counts])
    ax.set_xticks(np.arange(len(sorted_data_counts)))    
    fig.autofmt_xdate()
    
    plt.xlabel('Type of Information')
    plt.ylabel('Percent Likelihood of Access')
    plt.title('When Can a %s Access Your Information?' % title)
    remove_border()
    
    if filename:
        plt.savefig('images/%s.png' % filename)
```

```python
plot_data_percents(stranger_data, 'Stranger', filename='img_1')
```

```python
plot_data_percents(mutual_friends_data, 'Mutual Friend', filename='img_2')
```

```python
plot_data_percents(friends_data, 'Friend', filename='img_3')
```

```python
"""
Function
--------
levels_of_access_plot

Plots number of individuals willing to share each type of information
at each level of access

Parameters
----------
strangers : DataFrame
    Data available to strangers
mutual_friends : DataFrame
    Data available to mutual friends
friends : DataFrame
    Data available to friends
title : string
    The title information for the plot
"""
def levels_of_access_plot(strangers, mutual_friends, friends, filename=None):
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    
    data_counts = [(name, 
                    len(strangers[strangers[name] != 'Unlisted']),
                    len(mutual_friends[mutual_friends[name] != 'Unlisted']),
                    len(friends[friends[name] != 'Unlisted'])) for name in mutual_friends.columns]
    sorted_data_counts = sorted(data_counts, key=operator.itemgetter(1))
    
    total = len(mutual_friends_data)
    x_pos = np.arange(len(sorted_data_counts))
    
    ax.bar(x_pos,
           [item[1] for item in sorted_data_counts],
           align='center', color=dark, edgecolor='white', linewidth=5)
    ax.bar(x_pos,
           [item[2] - item[1] for item in sorted_data_counts],
           align='center', color=medium, edgecolor='white', linewidth=5,
           bottom=[item[1] for item in sorted_data_counts])
    ax.bar(x_pos,
           [item[3] - item[2] for item in sorted_data_counts],
           align='center', color=light, edgecolor='white', linewidth=5,
           bottom=[item[2] for item in sorted_data_counts])
    ax.bar(x_pos,
           [total - item[3] for item in sorted_data_counts],
           align='center', color=filler, edgecolor='white', linewidth=5,
           bottom=[item[3] for item in sorted_data_counts])
    
    ax.set_xticklabels([item[0] for item in sorted_data_counts])
    ax.set_xticks(x_pos)
    fig.autofmt_xdate()
    
    plt.xlabel('Type of Information')
    plt.ylabel('Number of People Who Allow Access')
    plt.title('Levels of Access')
    plt.legend(['For Strangers', 'For Mutual Friends', 'For Friends', 'For No One'], frameon=False,
               bbox_to_anchor=(0., -0.3, 1., .102), loc='lower center', fontsize=20, ncol=4)
    
    remove_border()
    
    if filename:
        plt.savefig('images/%s.png' % filename)
```

```python
levels_of_access_plot(stranger_subset, mutual_friends_data, friends_data, 'img_4')
```

```python
"""
Function
--------
make_photo_df

Extract more meaningful photo-sharing information

Parameters
----------
data : DataFrame
    The data to use

Returns
----------
result : DataFrame
    A new DataFrame that counts different levels of access to photos
    rather than the more general information included initially
"""
def make_photo_df(data):
    result = pd.DataFrame(columns=['Subset of Cover', 'Subset of Profile', 'All Cover', 'All Profile', 'All Photos'],
                          index=data.index)
    result = result.fillna('Unlisted')
        
    for row in data.index:
        if (re.search(r"\w*subset\w*", data.ix[row, 'Photos'])) and (re.search(r"\w*cover\w*", data.ix[row, 'Photos']) or re.search(r"\w*Cover\w*", data.ix[row, 'Photos'])):
            result.ix[row, 'Subset of Cover'] = 1
        if (not re.search(r"\w*subset\w*", data.ix[row, 'Photos'])) and (re.search(r"\w*cover\w*", data.ix[row, 'Photos']) or re.search(r"\w*Cover\w*", data.ix[row, 'Photos'])):
            result.ix[row, 'Subset of Cover'] = 1
            result.ix[row, 'All Cover'] = 1
        
        if (re.search(r"\w*subset\w*", data.ix[row, 'Photos'])) and (re.search(r"\w*profile\spics\w*", data.ix[row, 'Photos'])):
            result.ix[row, 'Subset of Profile'] = 1
        if (not re.search(r"\w*subset\w*", data.ix[row, 'Photos'])) and (re.search(r"\w*profile\spics\w*", data.ix[row, 'Photos'])):
            result.ix[row, 'Subset of Profile'] = 1
            result.ix[row, 'All Profile'] = 1
        
        if (data.ix[row, 'Photos'].strip() == 'full pictures'):
            result.ix[row, 'Subset of Cover'] = 1
            result.ix[row, 'All Cover'] = 1
            result.ix[row, 'Subset of Profile'] = 1
            result.ix[row, 'All Profile'] = 1
            result.ix[row, 'All Photos'] = 1
            
    return result

strangers_photo_data = make_photo_df(stranger_subset)
mutual_friends_photo_data = make_photo_df(mutual_friends_data)
friends_photo_data = make_photo_df(friends_data)

levels_of_access_plot(strangers_photo_data, mutual_friends_photo_data, friends_photo_data, filename='img_11')
```

```python
"""
Function
--------
plot_delta_count

Plots number of individuals willing to share each type of information
at two levels of access

Parameters
----------
data1 : DataFrame
    Data available to level 1 of access
data2 : DataFrame
    Data available to level 1 of access
title1 : string
    The title information for the plot for level 1 of access
title2 : string
    The title information for the plot for level 2 of access
"""
def plot_delta_count(data1, data2, title1, title2, filename=None):
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    
    data_counts = [(name, 
                    len(data1[data1[name] != 'Unlisted']),
                    len(data2[data2[name] != 'Unlisted'])) for name in data1.columns]
    sorted_data_counts = sorted(data_counts, key=operator.itemgetter(1))
    
    total = len(data1.index)
    x_pos = np.arange(len(sorted_data_counts))
    
    ax.bar(x_pos,
           [item[1] for item in sorted_data_counts],
           align='center', color=dark, edgecolor='white', linewidth=5)
    ax.bar(x_pos,
           [item[2] - item[1] for item in sorted_data_counts],
           align='center', color=light, edgecolor='white', linewidth=5,
           bottom=[item[1] for item in sorted_data_counts])
    ax.bar(x_pos,
           [total - item[2] for item in sorted_data_counts],
           align='center', color=filler, edgecolor='white', linewidth=5,
           bottom=[item[2] for item in sorted_data_counts])
    
    ax.set_xticklabels([item[0] for item in sorted_data_counts])
    ax.set_xticks(x_pos)
    fig.autofmt_xdate()
    
    plt.xlabel('Type of Information')
    plt.ylabel('Access Change, %s to %s' % (title1, title2))
    plt.title('From %s to %s' % (title1, title2))
    plt.legend(['For %ss' % title1, 'For %ss' % title2, 'For Neither'], frameon=False,
           bbox_to_anchor=(0., -0.30, 1., .102), loc='lower center', fontsize=20, ncol=4)
    
    remove_border()
    
    if filename:
        plt.savefig('images/%s.png' % filename)
```

```python
plot_delta_count(stranger_subset, mutual_friends_data, 'Stranger', 'Mutual Friend', filename='img_8')
```

```python
plot_delta_count(mutual_friends_data, friends_data, 'Mutual Friend', 'Friend', filename='img_9')
```

```python
plot_delta_count(stranger_subset, friends_data, 'Stranger', 'Friend', filename='img_10')
```
